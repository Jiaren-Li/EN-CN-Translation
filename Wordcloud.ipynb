```python
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
from PIL import Image
import matplotlib.pyplot as plt
import jieba
import numpy as np
```

```python
text_en=open('train.en','r',encoding='utf-8').read()
text_zh=open('train.zh','r',encoding='utf-8').read()
```

```python
wc=WordCloud(mode='RGBA',max_font_size=2000,background_color=None,collocations=False)
wc.generate(text_en)

plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc=WordCloud(mode='RGBA',font_path='Dengb.ttf',max_font_size=2000,background_color=None,collocations=False)
wc.generate(text_zh)
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```
